```python
import numpy as np
cf1 = [-10000, 6500, 3000, 3000, 1000]
cf2 = [-10000, 3500, 3500, 3500, 3500]

cf_test = [1000, 500, 400, 300, 100]
```

```python
import numpy_financial as npf
```

```python
cf_test_mirr = npf.mirr(cf_test, finance_rate=0.1, reinvest_rate=0.1)
#finance rate for negative outflows, reinvestment rate for positive inflows
print(f'MIRR for cf_test: {cf_test_mirr:.4f}')
```

```python
mirr1 = npf.mirr(cf1, finance_rate=0.12, reinvest_rate=0.12)
mirr2 = npf.mirr(cf2, finance_rate=0.12, reinvest_rate=0.12)
print(f'MIRR for cf1: {mirr1:.4f}')
print(f'MIRR for cf2: {mirr2:.4f}')
```

```python
def tv(cf, rate):
    sum = 0
    for i in range(len(cf)):
        if(cf[i] > 0):
            sum += cf[i] * ((1 + rate) ** (len(cf) - i-1))
    return sum
```

```python
print(f'TV for cf1: {tv(cf2, 0.12):.2f}')
```
