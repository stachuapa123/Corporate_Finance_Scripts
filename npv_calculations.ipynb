```python
def npv(i, cash_flows):
    return sum([cf / (1 + i) ** t for t, cf in enumerate(cash_flows)])
```

```python
cash_flows = [-1000, 200, 300, 400, 500]
discount_rate = 0.1

print(list(enumerate(cash_flows)))
```

```python
print(npv(0.11, cash_flows))
```

```python
def line(x1, y1, x2, y2):
    a = (y2 - y1)/(x2 - x1)
    b = y1 - a * x1
    return -b/(a)


def secand(l, r, k, Err, f):
    ct=1
    y1 = f(l) - k
    y2 = f(r) - k
    xs = line(l, y1, r, y2)
    vec = []
    while( abs(y2) > Err):
        #y1 = f(l) - k
        y2 = f(r) - k
        xs = line(l, y1, r, y2)
        ys = f(xs) - k
        vec.append(abs(ys))
        print(ct, "secand current root value: ", xs, "  error value: ", abs(ys))
        ct+=1
        l = r
        r = xs
        y1 = y2
    return xs,ct, vec

def fun(x):
    return npv(x, cash_flows=cash_flows)

k = 0 #rigth side = k
l = 0
r = 1
Err = 1e-3
rt, ii, v2 = secand(l, r, k, Err, fun)
```

```python
pa = [-3000, 1500, 1500, 1500, 1500]
pb = [-3000, 1000, 1000, 1000, 1000, 1000, 1000, 1000, 1000]
```

```python
npv_a = npv(0.12, pa)
npv_b = npv(0.12, pb)
print("NPV of project A: ", npv_a)
print("NPV of project B: ", npv_b)
```

```python
#chain replacement
n1 = len(pa)
n2 = len(pb)
print("Project A has ", n1, " cash flows.")
print("Project B has ", n2, " cash flows.")
npv_ac = npv_a * n2/n1
npv_bc = npv_b 
print("Chain replacement NPV of project A: ", npv_ac)
print("Chain replacement NPV of project B: ", npv_bc)
```

```python
i = 0.12
#equicalent annuity
ea1 = npv_a / sum([1/(1+0.12)**t for t in range(1, n1+1)])
ea1b = npv_a * ( i  / (1 - (1+i) ** (-n1))  )   # function for ea
ea2 = npv_b / sum([1/(1+0.12)**t for t in range(1, n2+1)])
ea2b = npv_b * ( i  / (1 - (1+i) ** (-n2))  )   # function for ea
print("Equivalent annuity of project A: ", ea1b)
print("Equivalent annuity of project B: ", ea2b)
```

```python
#infinite npv
def infinite_npv(cash_flow, i, npv_func=npv):
    NPV = npv_func(i, cash_flow)
    print("NPV for infinite NPV calculation: ", NPV)
    n = len(cash_flow)
    exp = (1+i)**n
    return NPV * exp / (exp - 1)

inf1 = infinite_npv(pa, 0.12)
inf2 = infinite_npv(pb, 0.12)

print("Infinite NPV of project A: ", inf1)
print("Infinite NPV of project B: ", inf2)
```
