```python
N_years = 18 
equity = 250000
end = equity * 4
```

```python
def FV(eq, years, interest, belka = False,  endval = 0):
    fv = eq * (1 + interest) ** years
    gowno = 0
    if(belka):
        gowno = (fv - eq) * 0.19
    return fv - endval - gowno
```

```python
def FVA(pmt, years, interest, belka = False):
    if(interest == 0):
        return pmt * years
    fva = pmt  * ( ((1 + interest) ** years - 1 ) / interest)
    gowno = 0
    if(belka):
        gowno = (fva - pmt * years) * 0.19
    return fva - gowno
```

```python
def FVAdue(pmt, years, interest, belka = False):
    if(interest == 0):
        return pmt * years
    fva = pmt  * (1 + interest) * ( ((1 + interest) ** years - 1 ) / interest)
    gowno = 0
    if(belka):
        gowno = (fva - pmt * years) * 0.19
    return fva - gowno
```

```python
def PVA(pmt, years, interest, belka = False):
    if(interest == 0):
        return pmt * years
    fva = pmt  *  ( (1 - (1 + interest) ** (-1 *years))  / interest)
    gowno = 0
    if(belka):
        gowno = (fva - pmt * years) * 0.19
    return fva - gowno

def PVAdue(pmt, years, interest, belka = False):
    if(interest == 0):
        return pmt * years
    fva = pmt * (1+interest) *  ( (1 - (1 + interest) ** (-1 *years))  / interest)
    gowno = 0
    if(belka):
        gowno = (fva - pmt * years) * 0.19
    return fva - gowno
```

```python
def line(x1, y1, x2, y2):
    a = (y2 - y1)/(x2 - x1)
    b = y1 - a * x1
    return -b/(a)


def secand(l, r, k, Err, f):
    ct=1
    y1 = f(l) - k
    y2 = f(r) - k
    xs = line(l, y1, r, y2)
    vec = []
    while( abs(y2) > Err):
        #y1 = f(l) - k
        y2 = f(r) - k
        xs = line(l, y1, r, y2)
        ys = f(xs) - k
        vec.append(abs(ys))
        print(ct, "secand current root value: ", xs, "  error value: ", abs(ys))
        ct+=1
        l = r
        r = xs
        y1 = y2
    return xs,ct, vec

def fun(x):
    return FV(eq = equity, years=N_years, interest=x, belka = True,  endval = 0)
```

```python
k = end # parameter k, in this situation we calculate 3rd degree root of this number
l = 0
r = 1
Err = 1e-3
rt, ii, v2 = secand(l, r, k, Err, fun)
```
